# Extracting categorical and numerical data

```python
def cat_num_splitter(data):
    cols = list(data.columns)
    cat_cols = []
    num_cols = []
    for col in cols:
        if data[col].dtype == 'int64' or data[col].dtype == 'int32' or data[col].dtype == 'float64':
            num_cols.append(col)
        else:
            cat_cols.append(col)

    cat_df = data[cat_cols]
    num_df = data[num_cols]

    # returns categorical and numerical data
    return cat_df, num_df
```

```python
# creating a function that will convert all height instances to 'cm' while removint the unit and converting the data type to float
def height_conv(data):
    # Initializing an empty list to store the converted height values
    hgt_list = []
    
    # Iterating over each height value in the 'Height' column of the categorical DataFrame
    for hgt in data:
        
        # Checking if the height value is already in centimeters
        if 'cm' in hgt:
            # Splitting the height value by 'cm' to isolate the numerical part
            hgt_split = hgt.split('cm')
            # Converting the numerical part to a float and appending it to the hgt_list
            hgt_list.append(float(hgt_split[0]))
        
        # If the height value is not in centimeters (implying it's in feet and inches)
        elif 'cm' not in hgt:
            # Splitting the height value by the apostrophe to separate feet from inches
            hgt_split = hgt.split("'")
            # Isolating the inches part and splitting it by the quotation mark to remove it
            hgt_decimal = hgt_split[1]
            hgt_decimal = hgt_decimal.split('"')
            # Converting the inches part to a float and dividing by 12 to convert it to feet
            hgt_inch = float(hgt_decimal[0])/12
            # Adding the feet part and the converted inches part to get the total height in feet
            hgt_feet = float(hgt_split[0]) + hgt_inch
            # Converting the total height from feet to centimeters (1 foot = 30.48 cm)
            hgt_cm = hgt_feet * 30.48
            # Appending the converted height in centimeters to the hgt_list
            hgt_list.append(hgt_cm)
    return hgt_list

# Adding a new column 'New_height' to the categorical DataFrame and setting its values to the converted heights
categorical_df['New_height'] = height_conv(categorical_df.Height)
```

```python
def weight_conv(data):
    # converting the weight measurement unit from pounds to kilogram  
    # 2.205 lbs = 1 kg
    New_weight = []
    
    # some of the instances in the data are of float datatype
    # cast the entire 'Weight' feature as string
    for weg in data.astype(str):
        
        # Check if the weight value is already in kilograms
        if 'kg' in weg:
            # Remove the 'kg' suffix from the weight value
            weg = weg[:-2]
            # Convert the remaining string to a float and append it to New_weight
            New_weight.append(float(weg))
        
        # If the weight value is in pounds
        elif 'lbs' in weg:
            # Remove the 'lbs' suffix from the weight value
            weg = weg[:-3]
            # Convert the remaining string to a float
            weg_lbs = float(weg)
            # Convert the weight from pounds to kilograms
            weg_kg = weg_lbs / 2.205
            # Append the converted weight to New_weight
            New_weight.append(weg_kg)
        
        # If the weight value is neither in k
        else:
            # Convert the string directly to a float and append it to New_weight
            New_weight.append(float(weg))
    return New_weight

# Add the new weight values as a column 'New_weight' to the categorical DataFrame
categorical_df['New_weight'] = weight_conv(categorical_df.Weight)
```

```python
# Define a function named 'value_converter' that takes a DataFrame column as input
def value_converter(data):
    # Initialize an empty list to store the converted values
    new_value = []
    
    # Iterate over each value in the 'Value' column of the DataFrame, casting each value to a string
    for val in data.astype(str):
        
        # Check if the value contains both the Euro symbol (€) and 'M' (indicating millions)
        if '€' and 'M' in val:
            # Convert the substring (removing the Euro symbol and 'M') to a float
            val = float(val[1:-1])
            # Append the converted value to the new_value list
            new_value.append(val)
        
        # Check if the value contains both the Euro symbol (€) and 'K' (indicating thousands)
        elif '€' and 'K' in val:
            # Convert the substring (removing the Euro symbol and 'K') to a float
            val = float(val[1:-1])
            # Convert the value from thousands to millions by dividing by 1,000
            val_m = val / 1000
            # Append the converted value to the new_value list
            new_value.append(val_m)
        
        # Check if the value contains only the Euro symbol (€) (implying it is in whole Euros)
        elif '€' in val:
            # Convert the substring (removing the Euro symbol) to a float
            val = float(val[1:])
            # Convert the value from Euros to millions by dividing by 1,000,000
            val = val / 1000000
            # Append the converted value to the new_value list
            new_value.append(val)
    
    # Return the list of converted values
    return new_value
categorical_df['new_value'] = value_converter(categorical_df.Value)
```

```python
# Define a function that converts object-type samples, handling NaN values appropriately
def datatype_converter(data):
    
    new_val = []  # Initialize an empty list to store the converted values

    # Iterate over each value in the data column, converting each to a string
    for i in data.astype(str):
        
        # Check if the value does not contain an underscore ('_')
        if '_' not in i:
            # Convert the value to a float
            i = float(i)
            # Append the converted float to the new_val list
            new_val.append(i)
        else:
            # If the value contains an underscore, remove the last character (assumed to be '_')
            i = float(i[:-1])
            # Append the modified and converted float to the new_val list
            new_val.append(i)

    # Return the list of converted values
    return new_val
```

```python
# Define a function to convert ratings that may contain stars to integer values
def ratn_converter(data):
    # Initialize an empty list to store the converted ratings
    new_val = []
    
    # Iterate over each rating in the data, converting each to a string
    for ratn in data.astype(str):
        
        # Check if the rating contains a star symbol ('★')
        if '★' in ratn:
            # Remove the star symbol and convert the remaining part to an integer
            ratn = int(ratn[:-1])
            # Append the converted integer rating to the new_val list
            new_val.append(ratn)
        else:
            # If the rating does not contain a star, append it as is
            new_val.append(ratn)

    # Return the list of converted ratings
    return new_val
```

After examining the features in numerical_df, it's evident that there are outliers present. I will write a function to analyze the percentage of outliers in each feature using the interquartile range (IQR).

```python
# Function to calculate the interquartile range (IQR) and the upper and lower limits for outliers
def in_quantile_rng(col):
    # Calculate the 75th percentile (Q3) and the 25th percentile (Q1) of the input column
    q3, q1 = np.percentile(col, [75, 25])
    
    # Compute the interquartile range (IQR) as the difference between Q3 and Q1
    iqr = q3 - q1
    
    # Calculate the upper limit for outliers (Q3 + 1.5 * IQR)
    upper_limit = q3 + 1.5 * iqr
    
    # Calculate the lower limit for outliers (Q1 - 1.5 * IQR)
    lower_limit = q1 - 1.5 * iqr
    
    # Return the upper and lower limits
    return upper_limit, lower_limit
```

```python
# Inspect the numerical_df DataFrame for outlier statistics
for col in numerical_df.columns:  # Iterate through each column in the numerical DataFrame
    print('-------------------------------------------------------')
    print(f'Outlier statistics for {col} column')  # Print the name of the current column

    # Calculate the upper and lower limits for outliers using the in_quantile_rng function
    ul, ll = in_quantile_rng(numerical_df[col])

    print(f'Upper limit is {ul}')
    print(f'Lower limit is {ll}')

    # Count the total number of outliers above the upper limit and below the lower limit
    total_outliers = len(numerical_df.loc[numerical_df[col] < ll, col]) + len(numerical_df.loc[numerical_df[col] > ul, col])

    # Calculate the percentage of outliers relative to the total number of entries in the DataFrame
    percent_outliers = (total_outliers / len(numerical_df.index)) * 100

    # Print the percentage of outliers in the current column
    print(f'Percentage of outliers in {col} is {percent_outliers}')

    # Check if the percentage of outliers exceeds 5%
    if percent_outliers > 5:
        print(f'Heavy presence of outliers in {col}')  # Print a warning if outliers are significant

    print('-------------------------------------------------------')
```

Based on the analysis, it is evident that only the 'GK Reflexes' feature has a significant number of outliers, accounting for approximately 10.98% of the data. This can be considered acceptable, as most players in the dataset are not goalkeepers.

### Feature Engineering

```python
# Function to separate a DataFrame into categorical and numerical data
def cat_num_splitter(data):
    # Get the list of all column names in the DataFrame
    cols = list(data.columns)
    
    # Initialize empty lists to store categorical and numerical column names
    cat_cols = []
    num_cols = []
    
    # Iterate through each column in the DataFrame
    for col in cols:
        # Check the data type of the column
        if data[col].dtype == 'int64' or data[col].dtype == 'int32' or data[col].dtype == 'float64':
            num_cols.append(col)  # If numerical, append to num_cols list
        else:
            cat_cols.append(col)  # If categorical, append to cat_cols list

    # Create a DataFrame for categorical columns
    cat_df = data[cat_cols]
    
    # Create a DataFrame for numerical columns
    num_df = data[num_cols]

    # Return the DataFrames containing categorical and numerical data
    return cat_df, num_df
```
